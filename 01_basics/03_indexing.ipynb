```python
import pandas as pd

df = pd.DataFrame({
    'product name': ['laptop', 'keyboard', 'monitor'],
    'price': [1260000.0, 47250.0, 200000],
    'stock': [10, 30, 20],
    'total': [12600000.0, 1417500.0, 4000000]
})

print(df)
```

```python
df[df['price'] > 100000] #Single condition format
```

```python
df[(df['price'] > 100000) & (df['stock'] < 30)] #multiple conditions format
```
