```python
import pandas as pd

data = {
    'name': ['Rachel', 'Smith', 'Micheal'],
    'age': [25, 30, 28],
    'city': ['London', 'Manchester', 'Brighton']
}

print(data)

df = pd.DataFrame(data)
print(df)
```

```python
df.shape        # (3, 3) → 3행 3열
df.columns      # Index(['이름', '나이', '도시'])
df.index        # RangeIndex(start=0, stop=3, step=1)
df.info()       # 열 타입, 결측치 등 기본 정보
df.describe()   # 수치형 열의 통계 요약
```

```python
df['name']             # return series
df[['name', 'city']]    # return DataFrame (multiple rows)

df['age'].mean()       # (calculate average)
```

```python
df.loc[0] #access column by index (label) 
df.iloc[1] #access colum by index (number)
```

## practice

```python
# data set

data = {
    'product name': ['laptop', 'mouse', 'keyboard'],
    'price': [1200000, 25000, 45000],
    'stock': [10, 50, 30]
}

df = pd.DataFrame(data)
print(df)
```

```python
df['price'].mean()
```

```python
df.loc[0, 'price']
```

### Add rows

```python
# editing column or add column
df['total'] = df['price'] * df['stock'] #create a new column named 'total' which has value of ('price' x 'stock') column
print(df)
```

### Overwrite existing column

```python
df.loc[3] = ['monitor', 200000, 20, 200000 * 20] #have to input correct index manually
print(df)
```

### Delete columns or rows

```python
df = df.drop(columns=['total']) #delete column
print(df)
```

```python
df = df.drop(index=1)  # delete index 1 row
print(df)
```
