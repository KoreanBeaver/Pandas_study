```python
import pandas as pd
#시리즈 생성
s = pd.Series([10, 20, 30, 40])
print(s) #출력값의 0 1 2 3 은 자동으로 부여되는 인덱스 값
```

```python
#시리즈 생성 및 커스텀 인덱스 값 부여
s2 = pd.Series([100, 200, 300], index=['a', 'b', 'c'])
print(s2)
```

```python
#인덱스를 통한 실습
print(s2['b'])
print(s2[0])
```

```python
print(s2.values)    # array([100, 200, 300])
print(s2.index)     # Index(['a', 'b', 'c'], dtype='object')
print(s2.mean())    # 평균값 계산: 200.0
```
